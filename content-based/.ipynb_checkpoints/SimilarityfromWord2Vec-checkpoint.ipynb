```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
#uploading word2vec model to extract products' features
model = Word2Vec.load("word2vec.model")
```

```python
from rake_nltk import Rake
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
#uploading preprocessed meta data
meta_df_reorganized=pd.read_csv('../model/meta_df_reorganized.csv',index_col=0)
meta_df_reorganized.head()
```

```python
#extracting key words of name column
meta_df_reorganized['Key_words'] = ''
r = Rake()
```

```python
for index, row in meta_df_reorganized.iterrows():
    r.extract_keywords_from_text(row['name'])
    key_words_dict_scores = r.get_word_degrees()
    row['Key_words'] = list(key_words_dict_scores.keys())
```

```python
#casting 'brand','category' and 'subcategory' columns to lowercase
meta_df_reorganized['brand'] = meta_df_reorganized['brand'].map(lambda x: x.split(','))
meta_df_reorganized['category'] = meta_df_reorganized['category'].map(lambda x: x.split(','))
meta_df_reorganized['subcategory'] = meta_df_reorganized['subcategory'].map(lambda x: x.split(','))
for index, row in meta_df_reorganized.iterrows():
    row['brand'] = [x.lower() for x in row['brand']]
    row['category'] = [x.lower() for x in row['category']]
    row['subcategory'] = [x.lower() for x in row['subcategory']]
```

```python
#concatenating 'brand', 'category', 'subcategory', 'Key_words' columns' unique elements to create bag of words for each product
meta_df_reorganized['Bag_of_words'] = ''
columns = ['brand', 'category', 'subcategory', 'Key_words']
for index, row in meta_df_reorganized.iterrows():
    words=[]
    for col in columns:
        words.extend(row[col])
    row['Bag_of_words'] = ' '.join(list(set(words)))
    
meta_df_reorganized = meta_df_reorganized[['productid','Bag_of_words']]
meta_df_reorganized.head()
```

```python
#starting feature extraction for each product
item_features=pd.DataFrame(columns=range(100))
```

```python


for index, row in meta_df_reorganized.iterrows():
    word_vectors = np.empty((0,100))
    words=row['Bag_of_words'].split()
    #by using wor2vec model, extracting vectors for each bag of words
    for w in words:
        if w in model.wv:
            word_vectors = np.append(word_vectors, np.array([model.wv[w]]), axis = 0)
        
    #an item's feature vector is the mean vector of its bag of words' vectors
    item_features.loc[row['productid']]=np.mean(word_vectors, axis = 0)
```

```python
#if there is a nan value in feature vector, filling it with 0
item_features=item_features.fillna(0)
```

```python
item_features = item_features.dropna(axis=1)
print(item_features.shape)
```

```python
#saving feature vector for API
item_features.to_csv('item_features.txt', sep='\t')
```

```python
#rereading to test nothing wrong with load item features
item_features = pd.read_csv('item_features.txt', sep='\t',index_col=0)
```

```python
item_features
```

```python
#calculating cosine similarity matriz for products with feature vector
from sklearn.metrics.pairwise import cosine_similarity
sg = cosine_similarity(item_features.values, item_features.values)
```

```python
sg
```

```python
#this similarity function will be used for creating recommendations
item_similarity=pd.DataFrame(sg, index=item_features.index, columns=item_features.index)
```

```python
item_similarity
```

```python
#lets test this similarity matrix with the test data that created for session-based recommender
events_df_reorganized=pd.read_csv('../events_df_reorganized_test.txt', sep='\t', dtype={'productid': str})

#For my evaluation,first of all, I will get the last added products of each session (these will be my targets)
events_df_reorganized=events_df_reorganized.loc[events_df_reorganized['productid'].isin(item_similarity.index)]
idx = events_df_reorganized.groupby(['sessionid'])['time'].transform(max) == events_df_reorganized['time']
test_dataset = events_df_reorganized[idx]
print(test_dataset.shape)
```

```python
#Remaining intances will be used for creating my recommendations
idx_train = events_df_reorganized.groupby(['sessionid'])['time'].transform(max) != events_df_reorganized['time']
train_dataset = events_df_reorganized[idx_train]
print(train_dataset.shape)
```

```python
#To make implementation much easier, I grouped the products according to their sessions
train_dataset=train_dataset.groupby('sessionid')['productid'].apply(list).reset_index()
train_dataset.head()
```

```python
test_dataset=test_dataset.groupby('sessionid')['productid'].apply(list).reset_index()
test_dataset.head()
```

```python
top_N = 10
rec_sum=0
for index, row in train_dataset.iterrows():
        #for a session getting current products' similarities with other products from similarity matrix
        partofsg=item_similarity.loc[row['productid']]
        
        #getting column indexes of top top_N + partofsg.shape[0](since there will be 1.) similarity scores
        top=top_N+partofsg.shape[0]
        idx = np.argpartition(partofsg.values, partofsg.values.size - top, axis=None)[-top:]
        result = np.column_stack(np.unravel_index(idx, partofsg.values.shape))
        indices=result[:,1]
        
        #getting the product ids of most similar(recommended) products
        recommends = [item_similarity.columns[i] for i in indices]
        
        #checking that whether one of the recommended item is bought in the session
        if test_dataset.iloc[index]['productid'][0] in recommends:
            rec_sum+=1
    
```

```python
#recall@10
#ratio of recommended and bough products and total number of session
float(rec_sum/train_dataset.shape[0])
```
