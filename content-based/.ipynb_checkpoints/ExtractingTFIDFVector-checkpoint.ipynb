```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import RegexpTokenizer
from rake_nltk import Rake
import re
import string
import random
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
meta_df_reorganized=pd.read_csv('../meta_df_reorganized.csv',index_col=0)
meta_df_reorganized=meta_df_reorganized.loc[meta_df_reorganized['category'] != 'Meyve ve Sebze']
meta_df_reorganized.head()
```

```python
meta_df_reorganized['Key_words'] = ''
r = Rake()
```

```python
for index, row in meta_df_reorganized.iterrows():
    r.extract_keywords_from_text(row['name'])
    key_words_dict_scores = r.get_word_degrees()
    row['Key_words'] = list(key_words_dict_scores.keys())
```

```python
meta_df_reorganized['brand'] = meta_df_reorganized['brand'].map(lambda x: x.split(','))
meta_df_reorganized['category'] = meta_df_reorganized['category'].map(lambda x: x.split(','))
meta_df_reorganized['subcategory'] = meta_df_reorganized['subcategory'].map(lambda x: x.split(','))
for index, row in meta_df_reorganized.iterrows():
    row['brand'] = [x.lower().replace(' ','') for x in row['brand']]
    row['category'] = [x.lower().replace(' ','') for x in row['category']]
    row['subcategory'] = [x.lower().replace(' ','') for x in row['subcategory']]
```

```python
meta_df_reorganized['Bag_of_words'] = ''
columns = ['brand', 'category', 'subcategory', 'Key_words']
for index, row in meta_df_reorganized.iterrows():
    words = ''
    for col in columns:
        words += ' '.join(row[col]) + ' '
    row['Bag_of_words'] = words
    
meta_df_reorganized = meta_df_reorganized[['productid','Bag_of_words']]
meta_df_reorganized.head()
```

```python
#Converting text descriptions into vectors using TF-IDF using Bigram
tf = TfidfVectorizer(ngram_range=(2, 2), stop_words='english', lowercase = False)
tfidf_matrix = tf.fit_transform(meta_df_reorganized['Bag_of_words'])
total_words = tfidf_matrix.sum(axis=0) 
#Finding the word frequency
freq = [(word, total_words[0, idx]) for word, idx in tf.vocabulary_.items()]
freq =sorted(freq, key = lambda x: x[1], reverse=True)
#converting into dataframe 
bigram = pd.DataFrame(freq)
bigram.rename(columns = {0:'bigram', 1: 'count'}, inplace = True) 
#Taking first 20 records
bigram = bigram.head(20)

#Plotting the bigram distribution
bigram.plot(x ='bigram', y='count', kind = 'bar', title = "Bigram disribution for the top 20 words in the bag of words", figsize = (15,7), )
```
